```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Goal: Classification of amino acids
properties = {
    "nonpolar": ["A", "V", "L", "I", "M", "P", "G", "W"],
    "polar": ["S", "T", "C", "N", "Q", "Y"],
    "aromatic": ["F", "Y", "W"],
    "basic": ["K", "R", "H"],
    "acidic": ["D", "E"]
}
# Print each property and the corresponding amino acids
for prop, aa_list in properties.items():
    aa_str = ", ".join(aa_list)
    print(f"{prop.upper()}: {aa_str}")

 # Create DataFrame for plotting
data = {"Property": [], "AminoAcid": []}
for prop, aa_list in properties.items():
    for aa in aa_list:
        data["Property"].append(prop)
        data["AminoAcid"].append(aa)

df = pd.DataFrame(data)

# Count occurrences per property
property_counts = df['Property'].value_counts().reset_index()
property_counts.columns = ['Property', 'Count']

# Plot
sns.set(style="whitegrid")
plt.figure(figsize=(8, 5))
sns.barplot(x="Property", y="Count", data=property_counts, palette="pastel")
plt.title("Amino Acids by Property")
plt.xlabel("Property")
plt.ylabel("Number of Amino Acids")
plt.tight_layout()
plt.show()
```
